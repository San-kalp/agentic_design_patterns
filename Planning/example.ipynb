```python
import os
from dotenv import load_dotenv
from crewai import Agent, Task, Crew, Process
from langchain_openai import ChatOpenAI
```

```python
load_dotenv(dotenv_path="/Users/sankalpchordia/code/agentic_design_patterns/Planning/.env")
```

```python
llm = ChatOpenAI(model= "gpt-4o-mini")
```

```python
llm.invoke("Hi")
```

```python
#Define agent 
planner_writer_agent = Agent(
    role="Article Planner and Writer",
    goal = 'Plan and then write concise, engaging summary on a specified topic.',
    backstory = ('You are an expert in writing technical content.' 'Your strenght lies in creating a clear, actionable plan before writing. ' 'Ensuring the final summary is both informative and easy to digest. '), # This is context about the agent's expertise and background,
    verbose = True, 
    allow_delegation = False,
    llm = llm 
)
```

```python
#Define task 
topic = "The importance of Reinforcement Learning in AI"
high_level_task = Task(
    description=(
        f"1. Create a bullet-point plan for a summary on the topic : '{topic}' \n"
        f"2. Write the summary based on your plan, keeping it around 200 words."
    ), 
    expected_output= (
        "A final report containing two distinct sections.\n\n"
        "### Plan \n"
        "- A bulleted list outlining the main points of the summary. \n\n"
        "### Summary\n"
        "- A concise and well strucutred summary of the topic."
    ),
    agent = planner_writer_agent
)
```

```python
crew = Crew(
    agents=[planner_writer_agent],
    tasks=[high_level_task],
    process= Process.sequential
)
```

```python
print("## Running the planning and writing task ##")
result = crew.kickoff()
```
