```python
import os
import asyncio
from typing import List 
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.agents import create_agent
from langchain_core.prompts import ChatPromptTemplate
from langchain.tools import tool
```

```python
load_dotenv()
```

```python
@tool
def search_information(query:str) :
    """Provide factual info on a given topic. Use this tool to find answers to phrases like 'what is the captial of France ?'"""
    print("Tool called")
    simulated_results = {
      "weather in london": "Current weather in London: 12°C, partly cloudy with light winds from the west at 15 km/h. Humidity at 65%.",
      "python list comprehension": "List comprehensions provide a concise way to create lists. Syntax: [expression for item in iterable if condition].",
      "best restaurants in new york": "Top rated: 1. Le Bernardin (French), 2. Eleven Madison Park (Contemporary), 3. Peter Luger (Steakhouse).",
      "how to tie a tie": "Windsor knot steps: 1. Cross wide end over narrow, 2. Bring wide end up through loop, 3. Wrap around, 4. Pull through and tighten.",
      "tesla stock price": "Tesla (TSLA) is currently trading at $248.50, up 2.3% from yesterday's close.",
      "calories in banana": "A medium banana (118g) contains approximately 105 calories, 27g carbs, 1.3g protein, and 0.4g fat.",
      "distance from earth to moon": "The average distance from Earth to the Moon is 384,400 km (238,855 miles).",
      "who invented the telephone": "Alexander Graham Bell is credited with inventing the first practical telephone in 1876.",
      "react usestate example": "const [count, setCount] = useState(0); - useState returns a state variable and a setter function.",
      "flight status aa123": "American Airlines Flight 123: On time. Departure: 10:30 AM JFK, Arrival: 1:45 PM LAX."
  }
    result = simulated_results.get(query.lower())
    return result
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature= 0,  top_p= 0.75, frequency_penalty=0.75)
tools   = [search_information]
```

```python
agent = create_agent(llm,tools,system_prompt="You are a helpful assistant")
```

```python
inputs = {"messages": [{"role": "user", "content": "weather in london"}]}
for chunk in agent.stream(inputs, stream_mode="updates"):
    print(chunk)
```
